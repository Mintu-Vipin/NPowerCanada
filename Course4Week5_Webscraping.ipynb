```python
# 4.	Write a Python program to extract and display all the image links from en.wikipedia.org/wiki/Peter_Jeffrey_(RAAF_officer)

import requests
from bs4 import BeautifulSoup
import re

url = 'https://en.wikipedia.org/wiki/Peter_Jeffrey_(RAAF_officer)'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

# Find all <img> tags within the content body of the page
content_body = soup.find(id='content')  # Assuming the main content has an id of 'content'
images = content_body.find_all('img', {'src': re.compile(r'.+\.jpg$|.png$|.jpeg$|.gif$')})

# Extract and display the URLs of the images
for image in images:
    # Get the 'src' attribute of the <img> tag
    image_url = image['src']

    # Construct the full URL if the src is relative
    if not image_url.startswith('http'):
        image_url = 'https:' + image_url

    # Display the image URL
    print(image_url)


    print()  # Print a new line for separation if needed
```

```python
# 8.	Write a Python program to display the contains of different attributes like different attributes like status_code, headers, url, history, encoding, reason, cookies, elapsed, request and content of a specified resource.

import requests
from bs4 import BeautifulSoup

def display_response_attributes(url):
    # Send a GET request to the specified URL
    response = requests.get(url)

    # Display status code
    print("Status Code:", response.status_code)

    # Display headers
    print("\nHeaders :", response.headers)

    # Display URL
    print("\nURL:", response.url)

    # Display redirection history (if any)
    if response.history:
        print("\nRedirection History:")
        for resp in response.history:
            print(f"{resp.status_code} - {resp.url}")

    # Display content encoding
    print("\nEncoding:", response.encoding)

    # Display reason phrase
    print("\nReason:", response.reason)

    # Display cookies
    print("\nCookies:")
    for cookie in response.cookies:
        print(cookie.name, ":", cookie.value)

    # Display elapsed time for request
    print("\nElapsed Time (in seconds):", response.elapsed.total_seconds())

    # Display request details (HTTP method and URL)
    print("\nRequest:")
    print(f"{response.request.method} {response.request.url}")

    # Display content (useful for textual responses like HTML)
    print("\nContent:")
    print(response.text)

# Example usage
url = 'https://en.wikipedia.org/wiki/Peter_Jeffrey_(RAAF_officer)'
display_response_attributes(url)
```

```python
# 8.	Write a Python program to remove the contents of a tag in a given html document.
from bs4 import BeautifulSoup

def remove_tag_contents(html_content, tag_name):
    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all instances of the specified tag
    tags = soup.find_all(tag_name)

    # Remove the contents of each found tag
    for tag in tags:
        # Clear the contents of the tag
        tag.clear()

    # Return the modified HTML content as a string
    return str(soup)

# Example HTML content
html_content = '''
<!DOCTYPE html>
<html>
<head>
    <title>Sample Page</title>
</head>
<body>
    <h1>Welcome to BeautifulSoup</h1>
    <p>This is a <strong>sample</strong> paragraph.</p>
    <div>
        <p>Another paragraph.</p>
        <a href="https://example.com">Link</a>
    </div>
</body>
</html>
'''

# Specify the tag name whose contents need to be removed
tag_name = 'p'

# Call the function to remove the contents of the specified tag
modified_html = remove_tag_contents(html_content, tag_name)

# Print the modified HTML content
print(modified_html)
```

```python
# 9.	Write a Python program to wrap an element in the specified tag and create the new wrapper.
from bs4 import BeautifulSoup

def wrap_element(html_content, target_tag, wrapper_tag):
    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the element(s) to be wrapped (in this case, all instances of the target_tag)
    elements = soup.find_all(target_tag)

    # Wrap each element with the specified wrapper_tag
    for element in elements:
        # Create a new wrapper tag
        wrapper = soup.new_tag(wrapper_tag)

        # Replace the element with the wrapper containing the element
        element.wrap(wrapper)

    # Return the modified HTML document as a string
    return str(soup)

# Example HTML content
html_content = '''
<!DOCTYPE html>
<html>
<head>
    <title>Sample Page</title>
</head>
<body>
    <h1>Welcome to Beautiful Soup</h1>
    <p>This is a sample paragraph.</p>
    <ul>
        <li>Item 1</li>
        <li>Item 2</li>
        <li>Item 3</li>
    </ul>
</body>
</html>
'''

# Specify the target tag and the wrapper tag
target_tag = 'p'
wrapper_tag = 'div'

# Call the function to wrap the target tag with the specified wrapper tag
modified_html = wrap_element(html_content, target_tag, wrapper_tag)

# Print the modified HTML document
print(modified_html)
```

```python
# 10.	Write a Python program to remove a tag from a given tree of html document and destroy it and its contents.
from bs4 import BeautifulSoup

def remove_tag_and_contents(html_content, tag_name):
    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all instances of the specified tag
    tags = soup.find_all(tag_name)

    # Remove each found tag along with its contents
    for tag in tags:
        tag.extract()  # Remove the tag and its contents from the parse tree

    # Return the modified HTML document as a string
    return str(soup)

# Example HTML content
html_content = '''
<!DOCTYPE html>
<html>
<head>
    <title>Sample Page</title>
</head>
<body>
    <h1>Welcome to Beautiful Soup</h1>
    <p>This is a <strong>sample</strong> paragraph.</p>
    <div>
        <p>Another paragraph inside a div.</p>
        <span>Some text in a span.</span>
    </div>
    <ul>
        <li>Item 1</li>
        <li>Item 2</li>
        <li>Item 3</li>
    </ul>
</body>
</html>
'''

# Specify the tag name to remove
tag_name = 'li'

# Call the function to remove the specified tag and its contents
modified_html = remove_tag_and_contents(html_content, tag_name)

# Print the modified HTML document
print(modified_html)
```
