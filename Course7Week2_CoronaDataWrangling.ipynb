```python
# 2. Write a python program to replace all the missing values with “Tobefound” for province column.
df['Province_State'].fillna("Tobefound", inplace = True)
df.isnull().sum()
```

```python
# 3. Normalize the values for “Confirmed” column as per the z-score.
mean = df['Confirmed'].mean()
median = df['Confirmed'].median()
df['Confirmed'].fillna((df['Confirmed']-mean)/median, inplace = True)
df.isnull().sum()
```

```python
# 4. Create bin values for “Death” column in the data set.
bins = np.linspace(min(df['Deaths']), max(df['Deaths']), 10)
```

```python
# 5. Write a Python program to get the maximum number of confirmed,
# deaths, recovered and active cases of Novel Coronavirus (COVID-19)
# Country wise.

df[['Confirmed', 'Recovered', 'Active']].groupby(df['Country_Region']).max()
```

```python
# 6. Write a Python program to get the total number of confirmed deaths
# and recovered people of Novel Coronavirus (COVID-19) cases
# Country/Region - Province/State wise

df['Total'] = df['Confirmed']+ df['Recovered']

# Group by 'Country_Region' and 'Province/State' and sum the total cases
df[['Total']].groupby(df['Country_Region']).max()
```
