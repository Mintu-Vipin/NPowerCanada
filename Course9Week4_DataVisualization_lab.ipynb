```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt


QUERY = """
SELECT ConvertedComp
FROM master
"""
df_hist = pd.read_sql_query(QUERY,conn)

# plot histogram
df_hist['ConvertedComp'].plot(kind='hist', figsize=(5, 3))
plt.title('Histogram of Converted Compensation')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age
FROM master
"""
df_Age = pd.read_sql_query(QUERY,conn)

df_Age['Age'].plot(kind='box', figsize=(5, 6))
plt.title("Box plots of Respondent's Age")

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df_scatter = pd.read_sql_query(QUERY,conn)

df_scatter.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(6,4), color='red')
plt.title('Relation Between Age and Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df_bubble = pd.read_sql_query(QUERY,conn)
df_bubble.plot(kind='scatter',
               x='WorkWeekHrs',
               y='CodeRevHrs',
               figsize=(10, 6),
               alpha=0.5,  # transparency
               color='pink'
               )

plt.title('Relation Between WorkWeekHrs and CodeRevHrs of Respondents')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.legend(['Age'], loc='upper right', fontsize='x-large')

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""
df_db = pd.read_sql_query(QUERY,conn)
df_groupby = df_db.groupby('DatabaseDesireNextYear', axis=0).count()
df_groupby.sort_values(["Respondent"], ascending=False, inplace=True)
df_pie = df_groupby.head()
df_pie['Respondent'].plot(kind='pie',
                          figsize=(5, 5),
                          autopct='%1.1f%%',
                          startangle=90,
                          shadow=True,
                          )

plt.title('Top 5 Databases that Respondents Wish to Learn Next Year')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT CodeRevHrs, WorkWeekHrs, Age
FROM master
"""
df_stacked = pd.read_sql_query(QUERY,conn)

df_stacked = df_stacked.loc[(df_stacked['Age'] >= 30) & (df_stacked['Age'] <= 35)].reset_index().drop(columns=['index'])
df_stacked.groupby(['CodeRevHrs','Age'])['WorkWeekHrs'].median().sort_values(ascending=False).unstack().plot(kind='bar', stacked=True, figsize=(20, 8))
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs For The Age Group 30 to 35', loc='center',fontsize=16, color='green')
plt.xlabel('CodeRevHrs', fontsize = 20)
plt.ylabel('WorkWeekHrs', fontsize = 20)
plt.legend(title='Age')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT ConvertedComp, Age
FROM master
"""
df_line = pd.read_sql_query(QUERY,conn)

df_line = df_line.loc[(df_line['Age'] >= 45) & (df_line['Age'] <= 60)].reset_index().drop(columns=['index'])
df_line = df_line.groupby('Age', axis=0).median()
df_line.plot(kind='line')

plt.title('Median ConvertedComp for all ages from 45 to 60.')
plt.ylabel('Converted Compensation')
plt.xlabel('Age')

plt.show()
```

```python
QUERY = """
SELECT ConvertedComp, Age
FROM master
"""
df_line1 = pd.read_sql_query(QUERY,conn)

df_line1 = df_line1.loc[(df_line1['Age'] >= 25) & (df_line1['Age'] <= 30)].reset_index().drop(columns=['index'])
df_line1 = df_line1.groupby('Age', axis=0).median()
df_line1.plot(kind='line')

plt.title('Median ConvertedComp for all ages from 45 to 60.')
plt.ylabel('Converted Compensation')
plt.xlabel('Age')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch, Respondent
FROM Master
"""
df_bar = pd.read_sql_query(QUERY,conn)

df_bar1 = df_bar.groupby('MainBranch', axis=0).count()

# generate plot
df_bar1.plot(kind='barh', figsize=(20, 10))
plt.xlabel('Total')
plt.title('Total Respondent Based on Main Branch')

plt.show()
```
